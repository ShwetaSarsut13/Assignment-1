1. What is the difference between a function and a method in Python?

In Python, the terms function and method both refer to callable blocks of code, but they have key differences in their context and usage:

1. Function:
Definition: A function is a standalone block of reusable code that performs a specific task.

Call: A function is called independently, and it can be invoked by its name.

Example:

```python
def my_function():
    print("This is a function.")

my_function()  # Call the function
```

2. Method:

Definition: A method is a function that is associated with an object (or class). It’s a function that is bound to an instance of a class and typically operates on the data contained in that instance.

Call: A method is called on an object, and it usually works on the object’s attributes or modifies the object’s state.

Example:


```python
class MyClass:
    def my_method(self):
        print("This is a method.")

obj = MyClass()
obj.my_method()  # Call the method on the object
```

Key Differences:

Association:

A function is not associated with any object unless passed as an argument.
A method is always associated with an object, typically using the self parameter.

Scope:

Functions can be defined and used anywhere in a Python script or module.
Methods are defined within classes and can only be used through an instance or the class itself (for class methods).

Syntax:

A function is called using its name, followed by parentheses.
A method is called on an object or class instance using the dot notation (object.method()).

2. Explain the concept of function arguments and parameters in Python.

In Python, parameters and arguments are terms used in the context of functions, and they refer to how data is passed to a function.

1. Parameters:

Definition: Parameters are the variables listed in a function’s definition. They act as placeholders that will hold the values passed to the function when it is called.

Example:

```python
def greet(name):  # 'name' is a parameter
    print(f"Hello, {name}!")
```

In this example, name is a parameter that will receive the actual value when the function is invoked.

2. Arguments:

Definition: Arguments are the actual values or data you pass to a function when you call it. These are assigned to the corresponding parameters in the function.

Example:

```python
greet("Alice")  # "Alice" is an argument
```

Here, "Alice" is the argument that gets passed to the name parameter of the greet function.

Types of Function Arguments in Python:

Python supports several ways to pass arguments to functions:

Positional Arguments:

Arguments are passed based on their position in the function call.

Example:

```python
def add(a, b):
    return a + b

result = add(3, 5)  # 3 is passed to 'a', 5 is passed to 'b'
```

Keyword Arguments:

You can explicitly specify which parameter to assign an argument to, regardless of the position.

Example:

```python
result = add(b=5, a=3)  # The order doesn't matter here
```

Default Arguments:

Parameters can have default values, so if no argument is provided, the default value is used.

Example:

```python
def greet(name="Guest"):
    print(f"Hello, {name}!")

greet()         # Prints "Hello, Guest!"
greet("Alice")  # Prints "Hello, Alice!"
```

Variable-Length Arguments:

These allow a function to accept an arbitrary number of arguments.

*args: For non-keyword variable-length arguments (positional)

```python
def multiply(*args):
    result = 1
    for num in args:
        result *= num
    return result

multiply(2, 3, 4)  # Returns 24
```

**kwargs: For keyword variable-length arguments.

```python
def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_info(name="Alice", age=30)  # Prints the keyword arguments
```

Summary of the Terms:

Parameters: Variables in the function definition that receive values when the function is called.

Arguments: The actual values that are passed to the function during a function call.

4. What is the purpose of the `return` statement in a Python function?

The return statement in a Python function is used to exit the function and send a value (or multiple values) back to the caller. It essentially determines the output of the function when it's called. The primary purposes of the return statement are:

1. Returning a Value to the Caller
The return statement allows a function to pass a result back to the place where it was called. Without return, the function will always return None by default.

Example:

```python
def add(a, b):
    return a + b  # This value is sent back to the caller

result = add(5, 3)  # result will be 8
```

2. Exiting the Function

The return statement also immediately terminates the function's execution. Once a return statement is encountered, the function stops running, and the control is passed back to the caller. Any code after the return statement will not be executed.

Example:

```python
def check_even(number):
    if number % 2 == 0:
        return True  # The function stops here if the number is even
    return False  # This line is only reached if the number is odd

print(check_even(4))
print(check_even(5))
```

3. Returning Multiple Values

Python functions can return multiple values in the form of a tuple. This is particularly useful when you want a function to compute and return more than one piece of data.

Example:

```python
def get_min_max(numbers):
    return min(numbers), max(numbers)

min_val, max_val = get_min_max([3, 1, 9, 5])
print(min_val)
print(max_val)
```

4. Returning Nothing (or Implicit None)

If a function doesn’t have a return statement, or the return statement is used without any value, the function will return None by default.

Example:

```python
def no_return():
    pass

print(no_return())
```

You can also explicitly use return without any value to exit the function and return None:

```python
def early_exit():
    return  # Returns None

print(early_exit())
```

5. Early Exit from a Function

The return statement can be used to exit a function early under certain conditions, especially when you only need the function to continue if certain criteria are met.

Example:

```python
def find_first_even(numbers):
    for num in numbers:
        if num % 2 == 0:
            return num  # Exit function when the first even number is found
    return None  # If no even number is found, return None

result = find_first_even([1, 3, 7, 4, 9])
print(result)
```

5. What are iterators in Python and how do they differ from iterables?

In Python, iterators and iterables are closely related concepts used to iterate over a sequence of data, but they have different roles. Let's break down both terms and their differences.

1. Iterable:
Definition: An iterable is any Python object capable of returning its members one at a time, allowing it to be iterated over (i.e., you can loop through it). Examples of iterables include lists, tuples, strings, dictionaries, and sets. In technical terms, an object is iterable if it implements the __iter__() method.

How It Works: When you use a for loop or other iteration tools, Python internally calls the __iter__() method of the iterable object, which returns an iterator.

Example of an Iterable:

```python
my_list = [1, 2, 3, 4]
for item in my_list:  # my_list is iterable
    print(item)
```

In this example, my_list is an iterable because you can loop through its elements.

2. Iterator:
Definition: An iterator is an object that represents a stream of data. It is the object that is returned by the __iter__() method of an iterable and provides a __next__() method, which returns the next element of the sequence. An iterator knows how to compute the next value and remembers the state (i.e., where it is in the sequence).

How It Works: When you call the next() function on an iterator, it retrieves the next item until there are no items left. At that point, it raises a StopIteration exception to signal the end of the iteration.

Example of an Iterator:

```python
my_list = [1, 2, 3, 4]
my_iterator = iter(my_list)  # Get the iterator from the iterable

print(next(my_iterator))  # Output: 1
print(next(my_iterator))  # Output: 2
print(next(my_iterator))  # Output: 3
print(next(my_iterator))  # Output: 4
# Calling next again will raise StopIteration
```

In this example, iter(my_list) returns an iterator object that can be used with next() to get each element.

Key Differences Between Iterables and Iterators:

1.Definition:

Iterable: An object that can return an iterator (i.e., can be iterated over).
Iterator: An object that produces items one at a time and represents a stream of data.

2.Methods:

Iterable: Must implement the __iter__() method, which returns an iterator.
Iterator: Must implement two methods:
__iter__() – returns the iterator itself.
__next__() – returns the next item in the sequence.

3.State:

Iterable: Does not maintain any state regarding iteration. Each time iter() is called, it returns a new iterator.
Iterator: Maintains state (it knows where it is in the sequence) and can only traverse the sequence once.

4. Reusability:

Iterable: You can loop over an iterable object (like a list) multiple times because calling iter() on it returns a fresh iterator each time.
Iterator: You can only iterate through an iterator once. After it is exhausted (i.e., it raises StopIteration), it cannot be reused unless a new iterator is created from the iterable.
Example to Illustrate the Difference:

```python
my_list = [1, 2, 3]

# 'my_list' is iterable
for item in my_list:
    print(item)

# 'iter(my_list)' returns an iterator
my_iterator = iter(my_list)

print(next(my_iterator))  # Output: 1
print(next(my_iterator))  # Output: 2

# Looping over the list again
for item in my_list:  # Fresh iterable, no issue
    print(item)

# Calling next on the same iterator after the first two items
print(next(my_iterator))  # Output: 3
```

Summary:

Iterable: An object you can iterate over (like lists, tuples, and strings).

Iterator: An object that produces one element at a time from an iterable, maintaining the current state of iteration.

Difference: An iterable returns an iterator, while an iterator is the actual object used to iterate through elements one by one using next().

6. Explain the concept of generators in Python and how they are defined.

In Python, generators are a special type of iterator that allows you to iterate over a sequence of values lazily (i.e., one at a time, on demand), which can be much more memory efficient than storing the entire sequence in memory at once. Generators are particularly useful when working with large datasets or infinite sequences.

Key Concepts of Generators:

1.Lazy Evaluation:

Generators don’t compute values all at once. Instead, they generate values on the fly and only when needed, making them more memory-efficient than normal functions that return lists.

2. Generator Functions:

A generator is defined like a normal function but uses the yield statement to return values one at a time, instead of return, which returns all at once and terminates the function.

The yield statement allows the generator to "pause" its execution and return a value, resuming where it left off the next time it is called.

3. Iterator Protocol:

Generators automatically implement Python’s iterator protocol (__iter__() and __next__() methods), so they can be used in for loops and other contexts that require an iterator.

When a generator's __next__() method is called (or when using next()), it resumes from the last yield statement, continues running until the next yield, and returns the next value.

Defining a Generator:
Syntax:
A generator is defined like a regular function, but instead of returning values, it yields them one at a time.

Example of a Simple Generator:

```python
def simple_generator():
    yield 1
    yield 2
    yield 3

# Calling the generator
gen = simple_generator()

# Iterating over the generator
print(next(gen))  # Output: 1
print(next(gen))  # Output: 2
print(next(gen))  # Output: 3
# Calling next again will raise StopIteration
```

In this example:

The function simple_generator() uses yield to produce values one at a time.
Each time next() is called, the generator resumes execution from the point where it last yielded a value.

Characteristics of Generators:

1.State Retention:

A generator retains its local variables and execution state between calls, allowing it to produce values one at a time without losing its place.

2.One-Time Iteration:

A generator can only be iterated through once. After it’s exhausted (i.e., it raises StopIteration), it cannot be reused unless a new generator is created.

3.Memory Efficiency:

Since generators produce items lazily (on-demand), they are memory efficient compared to lists or other data structures that store the entire sequence in memory.

Example: Generating a Range of Numbers:
Instead of using range(), which generates all the numbers at once, a generator can produce the numbers one by one.

```python
def count_up_to(max_value):
    current = 1
    while current <= max_value:
        yield current
        current += 1

counter = count_up_to(5)

for number in counter:
    print(number)
```

Output:

```python
1
2
3
4
5
```

Here, count_up_to(5) creates a generator that yields numbers from 1 to 5 lazily, one at a time.

Generator Expression:

In addition to generator functions, Python provides generator expressions, which are similar to list comprehensions but with parentheses instead of square brackets. Generator expressions are another way to create generators.

Example:

```python
gen_exp = (x * x for x in range(5))

for num in gen_exp:
    print(num)
```

Output:

```python
0
1
4
9
16
```

Here, gen_exp is a generator that yields the squares of numbers from 0 to 4, calculated lazily.

Benefits of Generators:

1.Memory Efficiency:

Generators are ideal for working with large datasets because they don’t require the entire data structure to be stored in memory. They only generate one item at a time, making them efficient for memory-constrained environments.

2.Infinite Sequences:

Generators can represent infinite sequences, which is impossible with lists since lists require storing all elements in memory.
Example: Infinite Fibonacci Sequence

```python
def fibonacci():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

fib = fibonacci()

for _ in range(10):
    print(next(fib))  # Output: First 10 Fibonacci numbers
```

3.Simple and Readable Code:

Generators provide a clean way to produce sequences without the overhead of managing state manually (like with an explicit iterator class).

Summary:

Generators are special types of iterators in Python, defined with yield instead of return.

They produce values lazily, making them memory-efficient.

They can be iterated over using for loops or next() and are useful for infinite sequences or large datasets.

Generator expressions provide a concise way to create generators using syntax similar to list comprehensions.

7. What are the advantages of using generators over regular functions?

Generators offer several advantages over regular functions, especially in scenarios where performance and memory efficiency are crucial. Below are the key benefits of using generators in Python:

1. Memory Efficiency:

Generators produce values one at a time and only when requested, which makes them much more memory-efficient than regular functions that return a full list of values.

In regular functions, when you return a list or another collection, the entire dataset is created in memory at once. For large datasets, this can lead to significant memory consumption.

Generators, on the other hand, generate items lazily, meaning they compute each item only when requested (e.g., using next()), thus avoiding the need to store the entire sequence in memory.

Example:



```python
def generate_numbers(limit):
    for i in range(limit):
        yield i

gen = generate_numbers(1000000)  # Generator takes very little memory
```

In contrast, if you were to return a list from a regular function, it would occupy significantly more memory:

```python
def generate_numbers_list(limit):
    return [i for i in range(limit)]

nums = generate_numbers_list(1000000)  # Memory-heavy operation
```

2. Performance Optimization (Faster Execution for Large Datasets):
Since generators calculate and return values on demand, they avoid the overhead of creating and storing large data structures all at once. This leads to faster execution when working with large datasets because Python doesn’t need to wait for the entire data structure to be built before starting to process it.

3. Ability to Handle Infinite Sequences:

Regular functions cannot handle infinite sequences because they would attempt to generate all the values at once, which would cause them to either run indefinitely or exhaust system memory.

Generators can easily work with infinite sequences, as they generate values on-the-fly, one at a time, and don’t need to know the size of the dataset beforehand.


Example: Infinite Fibonacci Sequence


```python
def fibonacci():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

fib = fibonacci()

for _ in range(10):
    print(next(fib))  # Generates Fibonacci numbers infinitely
```

4. Simpler Code for Iteration:

Generators simplify iteration logic by encapsulating the state of the iteration in a function with yield. In contrast, when using regular functions, you might need to manage state explicitly using a class or maintain complex loops.

Generators can be used in for loops and other constructs that accept iterators without needing to write custom iterator classes or functions.

5. State Retention:

With generators, you don't need to manage the state between successive calls. The generator automatically "remembers" the position where it last yielded a value and resumes from that point when the next value is requested.

In regular functions, keeping track of iteration state requires manual effort, such as using variables or maintaining the state in an external data structure (e.g., global variables or class attributes).


Example:



```python
def count_up_to(max_value):
    current = 1
    while current <= max_value:
        yield current
        current += 1

counter = count_up_to(5)
print(next(counter))  # Output: 1
print(next(counter))  # Output: 2
```

6. Pipeline Processing:

Generators can be used to create pipelines where data flows from one generator to another, processing each element lazily. This allows you to process large or even infinite streams of data incrementally, without having to generate or store all data at once.


Example: Processing a File Line by Line

```python
def read_large_file(file_path):
    with open(file_path) as file:
        for line in file:
            yield line

def process_lines(file_path):
    for line in read_large_file(file_path):
        if 'error' in line:
            yield line
```

In this example, generators enable reading and processing each line of a potentially large file without loading the entire file into memory at once.

7. Improved Performance in I/O-Bound Programs:

Generators allow for non-blocking or incremental processing of data, which is especially useful in I/O-bound programs (e.g., reading large files or streaming data from a network). Instead of waiting for the entire dataset to load, generators allow your program to start processing data immediately.

Example:

A generator can be used to download chunks of data from a network connection and process each chunk as it arrives, instead of waiting for the entire file to download.

8. Readable and Maintainable Code:

Using yield in generators can often simplify your code, making it easier to read and maintain. For example, without generators, you might have to write complex loops or maintain external state, whereas generators allow you to express the same logic concisely.

Example: Yielding Even Numbers

```python
def even_numbers(limit):
    for n in range(limit):
        if n % 2 == 0:
            yield n
```

Compared to a regular function returning a list of even numbers, this approach is much simpler and more efficient when the list is large.

Summary of Advantages:

1. Memory Efficiency: Generates items one at a time without creating the entire data structure at once.

2. Faster Execution: No need to wait for the full data structure to be built before processing.

3. Infinite Sequences: Can handle infinite data streams easily.

4. Simpler Iteration: Manages the iteration state internally with yield, simplifying code.

5. Pipeline Processing: Enables incremental processing of data streams, making it suitable for large-scale data or continuous input.

6. State Retention: Automatically resumes from where it last left off, retaining state between iterations.

7. Non-blocking I/O: Ideal for I/O-bound programs as it allows incremental processing.

In scenarios where you need to handle large datasets, avoid memory overhead, or process data incrementally, generators provide a more optimal and cleaner approach compared to regular functions that return collections (like lists).

8. What is a lambda function in Python and when is it typically used?

A lambda function in Python is a small, anonymous function defined using the lambda keyword. Unlike regular functions created using def, lambda functions are typically written in a single line and are used for short, simple operations where a full function definition might be overkill.

Syntax of a Lambda Function:








```python
lambda arguments: expression
```

arguments: The input(s) to the function, just like in a regular function.

expression: The operation or computation the function performs, and the result of this expression is returned automatically.

Characteristics of Lambda Functions:

Anonymous: Lambda functions are often called "anonymous" because they don't require a formal name like regular functions defined using def. However, they can be assigned to a variable if needed.

Single Expression: A lambda function can only contain a single expression. You can't include multiple statements, loops, or control structures like if, though you can use short conditional expressions (ternary operators).

Compact: Lambda functions are used when a quick, concise function is needed for a short task.

Example of a Lambda Function:

```python
add = lambda x, y: x + y
print(add(5, 3))  # Output: 8
```

Here, lambda x, y: x + y is a lambda function that takes two arguments (x and y) and returns their sum.

Use Cases for Lambda Functions:

Lambda functions are commonly used in situations where a simple function is needed for a short period of time, particularly in contexts where defining a full function using def would feel verbose. Below are some typical use cases:

1. Used with Functions Like map(), filter(), and reduce():
map(): Applies a function to each item in an iterable.

Example:

```python
numbers = [1, 2, 3, 4, 5]
squares = list(map(lambda x: x * x, numbers))
print(squares)  # Output: [1, 4, 9, 16, 25]
```

filter(): Filters items in an iterable based on a condition.

Example:

```python
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)  # Output: [2, 4]
```

reduce() (from functools module): Reduces a sequence to a single value based on a function.

Example:

```python
from functools import reduce
product = reduce(lambda x, y: x * y, numbers)
print(product)  # Output: 120
```

2. Sorting with key Parameter:

You can use lambda functions with the key argument in functions like sorted(), max(), and min() to specify a custom sorting order.

Example: Sorting by Second Element in a List of Tuples

```python
data = [(1, 'apple'), (3, 'banana'), (2, 'orange')]
sorted_data = sorted(data, key=lambda x: x[1])
print(sorted_data)  # Output: [(1, 'apple'), (3, 'banana'), (2, 'orange')]
```

Here, the list is sorted by the second element of each tuple ('apple', 'banana', 'orange').

3. In Inline Functions:

Lambda functions are often used when a function is needed only temporarily, like in cases where a full function definition would be cumbersome.

Example:

```python
button_click = lambda: print("Button clicked!")
button_click()  # Output: Button clicked!
```

4. Lambda with if-else (Ternary Conditionals):

A lambda function can contain conditional logic in a single line using a ternary if-else expression.

Example:

```python
max_val = lambda a, b: a if a > b else b
print(max_val(5, 10))  # Output: 10
```

This lambda function returns the greater of two numbers using a single-line conditional expression.

5. Lambda Inside Other Functions:

You can use lambda functions inside regular functions to define small, temporary operations that are only relevant within the context of the outer function.

Example:

```python
def multiplier(factor):
    return lambda x: x * factor

times_three = multiplier(3)
print(times_three(5))  # Output: 15
```

Here, the outer function multiplier returns a lambda function that multiplies the input by a factor.

Limitations of Lambda Functions:

Limited to a Single Expression: Lambda functions cannot have multiple expressions, statements, or complex logic like loops. For anything beyond a simple operation, a full function is better suited.

Reduced Readability: While lambda functions can make code concise, overusing them (especially with complex expressions) can lead to code that is harder to read and maintain.

No return Statement: In lambda functions, the result of the expression is automatically returned. You cannot explicitly use return.


Summary:
Lambda functions are small, anonymous functions defined using lambda and are typically used for short, simple tasks.

They are limited to a single expression, making them suitable for concise operations like mapping, filtering, or custom sorting.

Use cases include passing functions as arguments to map(), filter(), and sorted(), or defining temporary functions that don't need a formal name.

While powerful, lambda functions are not ideal for complex logic, as they can reduce code readability and are limited to single expressions.

9. Explain the purpose and usage of the `map()` function in Python.

The map() function in Python is used to apply a given function to every item of an iterable (such as a list, tuple, etc.) and return an iterator (or map object) with the results. It is useful when you want to transform or process all elements of an iterable in a concise way, without having to explicitly write loops.

Syntax of map():







```python
map(function, iterable[, iterable2, iterable3, ...])
```

function: The function that is applied to each item of the iterable. This function can be a built-in function, a user-defined function, or even a lambda function.

iterable: The iterable whose elements will be passed to the function. You can pass more than one iterable to map() if the function accepts multiple arguments.
How map() Works:

Single Iterable: If you pass a single iterable, map() applies the function to each element of that iterable.

Multiple Iterables: If you pass multiple iterables, map() applies the function to the corresponding elements of each iterable (pair-wise, or tuple-wise). The number of iterations is determined by the shortest iterable.


Example 1: Applying a Function to Each Element

10.  What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

The map(), reduce(), and filter() functions in Python are higher-order functions commonly used in functional programming to process and transform iterables (e.g., lists, tuples). While they have similar purposes, they differ in how they operate on data. Here’s a detailed comparison of these three functions:

1. map()

Purpose: map() applies a given function to each item in an iterable and returns an iterator (or map object) containing the results.
Use Case: It is used to transform or process every element in an iterable, like modifying or converting data.

Syntax:


```python
map(function, iterable)
```

Function: A function that takes one or more arguments and applies some operation to each element in the iterable(s).

Iterable: The iterable whose elements will be transformed (e.g., a list, tuple, etc.).

Example:

```python
numbers = [1, 2, 3, 4]
squared = map(lambda x: x * x, numbers)
print(list(squared))  # Output: [1, 4, 9, 16]
```

Here, map() applies the lambda function to square each number in the list.

Key Characteristics:

Works with one or more iterables.

Returns a map object (iterator), which can be converted to a list, tuple, etc.
Applies a function to every element of the iterable(s).

2. reduce()

Purpose: reduce() applies a function cumulatively to the items of an iterable, reducing it to a single value.

Use Case: It is used when you need to accumulate or aggregate values from a sequence to produce a single result (e.g., summing, multiplying, finding the greatest common divisor, etc.).

Note: reduce() is not a built-in function in Python 3 and needs to be imported from the functools module.


```python
from functools import reduce
reduce(function, iterable, initializer=None)
```

Function: A function that takes two arguments and applies an operation to the accumulated result and the next item.

Iterable: The iterable whose elements will be reduced.

Initializer (optional): A starting value for the accumulation process.

Example:


```python
from functools import reduce
numbers = [1, 2, 3, 4]
result = reduce(lambda x, y: x + y, numbers)
print(result)  # Output: 10
```

Here, reduce() accumulates the sum of all numbers in the list.

Key Characteristics:

Reduces the iterable to a single value.

Uses a binary function (a function that takes two arguments) to combine elements.

Returns a single output, which could be the sum, product, or some other aggregated result.


3. filter()

Purpose: filter() applies a given function to each item of an iterable and returns an iterator containing only the items for which the function returns True.

Use Case: It is used to filter out elements from an iterable based on a condition or predicate function (a function that returns a boolean value).

Syntax:

```python
filter(function, iterable)
```

Function: A function that returns True or False (a boolean function).

Iterable: The iterable whose elements are filtered.

Example:

```python
numbers = [1, 2, 3, 4, 5, 6]
evens = filter(lambda x: x % 2 == 0, numbers)
print(list(evens))  # Output: [2, 4, 6]
```

Here, filter() returns only the even numbers from the list.

Key Characteristics:

Returns an iterator containing only the filtered items that satisfy the condition.

The function should return a boolean (True or False) for each element.
Unlike map(), it does not modify the elements but instead selects them based on a condition.

When to Use:

Use map() when you need to apply the same function to each element of a sequence and return the transformed results.

Use reduce() when you need to combine all elements in a sequence into a single output through cumulative processing.

Use filter() when you need to remove elements that don't satisfy a certain condition and return only those that do.

Example Using All Three Functions Together:

```python
from functools import reduce

numbers = [1, 2, 3, 4, 5, 6]

# Using filter to get even numbers
evens = filter(lambda x: x % 2 == 0, numbers)

# Using map to square the even numbers
squared_evens = map(lambda x: x * x, evens)

# Using reduce to sum the squared even numbers
result = reduce(lambda x, y: x + y, squared_evens)

print(result)  # Output: 56 (since 2^2 + 4^2 + 6^2 = 4 + 16 + 36)
```

In this example:

filter() filters out the odd numbers.

map() squares the even numbers.

reduce() sums the squared even numbers.

By combining these functions, you can create powerful pipelines for data processing in Python!

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):  # Check divisibility up to the square root of n
        if n % i == 0:
            return False
    return True

# Example usage
number = 29
if is_prime(number):
    print(f"{number} is a prime number.")
else:
    print(f"{number} is not a prime number.")
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```python
class FibonacciIterator:
    def __init__(self, n_terms):
        self.n_terms = n_terms  # Total number of terms
        self.current_term = 0   # Track the current term number
        self.a, self.b = 0, 1   # Initialize the first two Fibonacci numbers

    def __iter__(self):
        return self  # Return the iterator object

    def __next__(self):
        if self.current_term >= self.n_terms:
            raise StopIteration  # Stop when the desired number of terms is reached
        if self.current_term == 0:
            self.current_term += 1
            return self.a  # Return the first term (0)
        elif self.current_term == 1:
            self.current_term += 1
            return self.b  # Return the second term (1)
        else:
            self.a, self.b = self.b, self.a + self.b  # Calculate next Fibonacci number
            self.current_term += 1
            return self.a  # Return the current Fibonacci number

# Example usage
fib = FibonacciIterator(10)  # Generate first 10 terms of Fibonacci sequence
for num in fib:
    print(num, end=" ")
```

Explanation:
Class Initialization (__init__):

n_terms: Specifies how many terms of the Fibonacci sequence you want.

self.a, self.b: These are the two starting numbers of the Fibonacci sequence (0 and 1).

self.current_term: Tracks how many terms have been generated.

__iter__(): This method returns the iterator object itself, which allows it to be used in a loop.

__next__(): This method defines how the next Fibonacci number is generated:

For the first two terms, it directly returns 0 and 1.

For subsequent terms, it calculates the next Fibonacci number by updating self.a and self.b (self.a gets the value of self.b, and self.b becomes the sum of the previous two numbers).

It stops the iteration when the specified number of terms is reached by raising
StopIteration.

10. Create a Python program that uses `filter()` to remove all the vowels from a given string

 Python program that uses the filter() function to remove all the vowels from a given string:










```python
# Function to check if a character is a vowel
def is_not_vowel(char):
    return char.lower() not in 'aeiou'

# Function to remove vowels from a string
def remove_vowels(input_string):
    # Using filter to apply the is_not_vowel function on each character
    filtered_chars = filter(is_not_vowel, input_string)
    return ''.join(filtered_chars)  # Joining the filtered characters back into a string

# Example usage
input_string = "Hello, World!"
result_string = remove_vowels(input_string)
print(f"String without vowels: {result_string}")
```

Explanation:

is_not_vowel(char): This helper function checks if a character is not a vowel. It converts the character to lowercase and checks if it is not in the string 'aeiou'.

remove_vowels(input_string): This function uses filter() to apply the is_not_vowel function to each character in the input string. It filters out the vowels, and then the remaining characters are joined back together into a new string using ''.join().

For the input string "Hello, World!", the output will be "Hll, Wrld!", with all vowels removed.
